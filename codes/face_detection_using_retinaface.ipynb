#### **Initialize the REtinaFace MobileNetV1 Model**

```python
detector = face_detection.build_detector("RetinaNetMobileNetV1", confidence_threshold=.5, nms_iou_threshold=.3)
```

```python

# Read image
image_path = "test_2.jpg"
img = cv2.imread(image_path)

# Save time
t0 = time.time()

# Getting the detections
detections = detector.detect(img)

# Calculate inference time
inf_time = round(time.time() - t0, 3)

# Print results
print(f"Detections: {detections}")
print(f"Inference time: {inf_time}s")
```

```python


# Draw detections
if len(detections) > 0:
    for detection in detections:
        # Converting predicted and ground truth bounding boxes to required format
        pred_bbox = detection

        pred_bbox[2] = pred_bbox[2] - pred_bbox[0]
        pred_bbox[3] = pred_bbox[3] - pred_bbox[1]

        pred_bbox = [int(i) for i in pred_bbox[:4]]

        cv2.rectangle(img,pred_bbox,(0,255,0),1)

# Write inference time
img = cv2.putText(img, f"Inf Time: {inf_time}s", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 1, cv2.LINE_AA)
     

# Display image
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
     
```

```python
import face_detection
import cv2
import time
import matplotlib.pyplot as plt

# Build the face detector
detector = face_detection.build_detector("RetinaNetResNet50", confidence_threshold=.5, nms_iou_threshold=.3)

# Read image
image_path = "test_2.jpg"
img = cv2.imread(image_path)

# Save time
t0 = time.time()

# Getting the detections
detections = detector.detect(img)

# Calculate inference time
inf_time = round(time.time() - t0, 3)

# Print results
print(f"Detections: {detections}")
print(f"Inference time: {inf_time}s")

# Counter for saved face images
face_counter = 0

# Draw detections and save individual faces
if len(detections) > 0:
    for detection in detections:
        # Extract bounding box coordinates
        x_min, y_min, x_max, y_max = detection[:4]
        
        # Convert to integer values
        x_min, y_min, x_max, y_max = int(x_min), int(y_min), int(x_max), int(y_max)
        
        # Crop the face from the image
        face = img[y_min:y_max, x_min:x_max]
        
        # Save the face as a separate image
        face_filename = f"face_{face_counter}.jpg"
        cv2.imwrite(face_filename, face)
        
        # Increment the counter
        face_counter += 1
        
        # Optionally display each face
        plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
        plt.title(f"Face {face_counter}")
        plt.show()
        
        # Draw bounding box on the original image
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (0, 255, 0), 3)

# Write inference time on the image
img = cv2.putText(img, f"Inf Time: {inf_time}s", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 3, cv2.LINE_AA)

# Convert the image from BGR to RGB for displaying with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image with bounding boxes
plt.imshow(img_rgb)
plt.title("Detected Faces")
plt.show()
```

## **Face Enhancement**

```python
from cv2 import dnn_superres
```

```python
# Create the SR object
sr = dnn_superres.DnnSuperResImpl_create()
```

```python
# Read Image
face_image_path = 'face_0.jpg'
face_image = cv2.imread(face_image_path)
```

```python
# Load the desired model
model_path = "LapSRN_x8.pb"
sr.readModel(model_path)
```

```python
# Set the desired model and scale to get correct pre- and post-processing
sr.setModel("lapsrn", 8)

# Upscale the image
result = sr.upsample(face_image)


# Display image
img = cv2.cvtColor(face_image,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

# Display image
img = cv2.cvtColor(result,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

# # Save the image
# cv2.imwrite("Monarch-butterfly-tall-l_upscaled_lapsrn.png", result)
```

```python
import numpy as np

kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp_1 = cv2.filter2D(src=result, ddepth=-1, kernel=kernel)

cv2.imshow('Sharpened Result image face', image_sharp_1)
cv2.waitKey()
cv2.destroyAllWindows()
```
